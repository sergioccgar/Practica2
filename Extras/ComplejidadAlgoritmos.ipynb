```python
import time
from random import randint
import matplotlib.pyplot as plt
import numpy as np
import math
```

```python
def constant_algo(items):
    result = items[0] * items[0]
    print(result)

constant_algo([4, 5, 6, 8])
```

```python
x = [2, 4, 6, 8, 10, 12]

y = [2, 2, 2, 2, 2, 2]

plt.plot(x, y, 'b')
plt.xlabel('Inputs')
plt.ylabel('Steps')
plt.title('Constant Complexity')
plt.show()
```

```python
def linear_algo(items):
    for item in items:
        print(item)

linear_algo([4, 5, 6, 8])
```

```python
x = [2, 4, 6, 8, 10, 12]

y = [2, 4, 6, 8, 10, 12]

plt.plot(x, y, 'b')
plt.xlabel('Inputs')
plt.ylabel('Steps')
plt.title('Linear Complexity')
plt.show()
```

```python
def linear_algo(items):
    for item in items:
        print(item)

    for item in items:
        print(item)

linear_algo([4, 5, 6, 8])
```

```python
x = [2, 4, 6, 8, 10, 12]

y = [4, 8, 12, 16, 20, 24]

plt.plot(x, y, 'b')

y = [elem*23 for elem in y]

plt.plot(x,y,'r')

plt.xlabel('Inputs')
plt.ylabel('Steps')
plt.title('Linear Complexity')
plt.show()
```

```python
def quadratic_algo(items):
    for item in items:
        for item2 in items:
            print(item, ' ' ,item)

quadratic_algo([4, 5, 6, 8])
```

```python
x = [2, 4, 6, 8, 10, 12]

y = [elem**2 for elem in x]

plt.plot(x, y, 'b')
plt.xlabel('Inputs')
plt.ylabel('Steps')
plt.title('Cuadratic Complexity')
plt.show()
```

```python
x = [2, 4, 6, 8, 10, 12]

y = [2, 2, 2, 2, 2, 2]

plt.plot(x, y, 'b')

y = [2, 4, 6, 8, 10, 12]
plt.plot(x,y,'r')

y = [elem*10000000 for elem in y]

plt.plot(x,y,'g')

y = [elem**8 for elem in x]
plt.plot(x,y,'black')

y = [math.factorial(elem) for elem in x]
plt.plot(x,y,'purple')

plt.xlabel('Inputs')
plt.ylabel('Steps')
plt.title('')
```
